```python
import pandas as pd
```

```python
companies = pd.read_csv("../../Datasets/Fortune 500/Fortune_500_Corporate_Headquarters_arcgis.csv")
#companies.head()
counties = pd.read_csv("../../Datasets/Housing Prices/County Datasets/All Homes County.csv")
ohcounties = counties[counties["State"] == "OH"]
ohcounties["RegionName"] = ohcounties.loc[:,"RegionName"].str.rstrip("County")
ohcounties = ohcounties.rename(columns={"RegionName":"index"})
ohcounties = ohcounties.set_index("index")
ohcounties = ohcounties.drop(ohcounties.loc[:, "RegionID":"SizeRank"], axis=1)
ohcounties = ohcounties.drop(ohcounties.loc[:, "RegionType":"MunicipalCodeFIPS"], axis=1)
ohcounties # should be 88
```

```python
ohcounties = ohcounties.transpose()
ohcounties = ohcounties.reset_index()
ohcounties = ohcounties.rename(columns={"index": "Date"})
ohcounties.to_excel("Ohio counties dataset.xlsx")
ohcounties
```
